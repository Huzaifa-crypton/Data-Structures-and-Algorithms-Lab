```python
import pandas as pd
from selenium import webdriver
from bs4 import BeautifulSoup
class Book:
    def __init__(self , bookName , pages , author):
        self.bookName = bookName
        self.pages = pages
        self.author = author

        
    
    
    
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Users\DR.NOOR KHAN\Week2\chromedriver.exe')
driver.get("https://www.scribd.com/books/categories")
content = driver.page_source
soup = BeautifulSoup(content)
```

```python
categoriesLink = []
for a in soup.findAll('h3' , attrs={'class':'_23qnUq'}):
    link = a.find('a',attrs={'class':'jShK3d'})['href']
    categoriesLink.append(link)
```

```python
allBooks = []
bnames = []
bpages = []
bauthors = []
i = 0
k = 0
for i in range (len(categoriesLink)):
    for j in range (1 , 11):
        url = "https://www.scribd.com"+categoriesLink[i]+"/all?page="+str(j)
        driver.get(url)
        content = driver.page_source
        soup = BeautifulSoup(content)
        for a in soup.findAll('div' , attrs={'class':'list_anchor_container'}):
            link = a.find('a',attrs={'class':'list_anchor book_link'})['href']
            book = a.find('div' , attrs = {"class":"title document_title"}).text
            author = a.find('span' , attrs = {"class":"author"}).text
            pages = a.find('div' , attrs = {"class":"read_time_estimate"}).text
            pages = pages.split()
            pages = pages[0].replace(',', '')
            b = Book(book , author , int(pages))
            print(b.bookName)
            print(b.pages)
            print(b.author)
            allBooks.append(b)
            bnames.append(book)
            bauthors.append(author)
            bpages.append(pages)
            #         driver.get(link)
            k+=1
            print(k)
#         content = driver.page_source
#         soup2 = BeautifulSoup(content)
        
#         for c in soup2.findAll('div' , attrs = {'class':"_9GlT8l"}):
#             i+=1
#             book = c.find('h1' , attrs = {"class":"FtRy_z"}).text
#             author = c.find('a' , attrs = {"class":"TfSX3o"}).text
#             pages = c.find('span' , attrs = {"data-e2e":"podcast-series-header-total-episodes"}).text
#             pages = pages.split()
#             pages = pages[0].replace(',', '')
#             b = Book(book , author , int(pages))
#             print(b.bookName)
#             print(b.pages)
#             print(b.author)
#             allBooks.append(b)
#             bnames.append(book)
#             bauthors.append(author)
#             bpages.append(pages)
        
print("YEASs")
df = pd.DataFrame({"BOOK NAME":bnames , "AUTHOR":bauthors , "PAGES":bpages})
df.to_csv("AllBooks.csv" , index = "True" , encoding = 'utf-8')

    
        
```
