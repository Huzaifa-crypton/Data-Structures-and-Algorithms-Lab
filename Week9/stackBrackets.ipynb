```python
class Node :
    key = None
    next = None
    def __init__(self,key):
        self.key = key 
    
class Stack:
    def __init__(self):
        self.head = None
        self.tail = None
        
    def push(self,z):
        n = Node(z)
        if (self.head == None):
            self.head = n
            self.tail = n
        else:
            self.tail.next = n
            self.tail = n
    def pop(self):
        y = self.tail
        temp = self.head
        if temp == y :
            self.head = None
            self.tail = None
            return temp
        while temp.next != y:
            temp = temp.next
        self.tail = temp
        self.tail.next = None
        
        return y
    def peek(self):
        if (self.head == None):
            return None
        return self.tail.key
            

exp = '[()'
s = Stack()
val = True
for char in exp:
    if char == "[" or char == "{" or char == "(" :
        s.push(char)
    else:
        brack = s.pop().key
        if brack == "(" and char != ")":
            val = False
        elif brack == "[" and char != "]":
            val = False
        elif brack == "(" and char != ")":
            val = False
num = s.peek()
if num != None:
    val = False
if val:
    print("BALNCED")
elif val == False:
    print("NOT BALANCE") 
```
